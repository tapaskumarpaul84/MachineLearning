```python
import pandas as pd
import numpy as np
```

```python
date_df['date']=pd.to_datetime(date_df['date'])
```

```python
date_df.info()
```

```python
date_df['year']=date_df['date'].dt.year
```

```python
date_df.sample(2)
```

```python
date_df['month of year']=date_df['date'].dt.month
```

```python
date_df.sample(2)
```

```python
date_df['month name of year']=date_df['date'].dt.month_name()
```

```python
date_df.sample(2)
```

```python
date_df['day']=date_df['date'].dt.day
date_df.sample(2)
```

```python
date_df['day of week']=date_df['date'].dt.dayofweek
date_df.sample(2)
```

```python
date_df['day of week name']=date_df['date'].dt.day_name()
date_df.sample(2)
```

```python
date_df['day weekend']=np.where(date_df['day of week name'].isin(['Sunday','Saturday']),1,0)
date_df.sample(5)
```

```python
date_df['week no']=date_df['date'].dt.isocalendar().week
date_df.sample(5)
```

```python
date_df['quarter no']=date_df['date'].dt.quarter
date_df.sample(5)
```

```python
date_df['semester no']=np.where(date_df['quarter no'].isin([1,2]),1,2)
date_df.sample(5)
```

```python
from datetime import datetime

now=datetime.now()
```

```python
date_df['past days']=(now-date_df['date']).dt.days
```
