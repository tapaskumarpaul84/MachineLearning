```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=2)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```

### Gradient Descent

```python
# consider m=93.15 and b=0 initially

m=93.15
b=0

y_pred=(m*x + b).reshape(4)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label="b=0")
plt.legend()
plt.show()
```

```python
x.ravel()
```

### Iteration-1

```python
m=93.15
b=0

loss_slope=-2* np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
# assuming learning rate lr=0.1

le=0.1

step_size=le * loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred_updated=(m*x + b).reshape(4)
y_pred_updated
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label="b=0")
plt.plot(x,y_pred_updated,color='yellow',label="b=updated")
plt.legend()
plt.show()
```

### Iteration-2

```python
## b=-26.88
loss_slope=-2* np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
step_size=le * loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred_itr2=(m*x + b).reshape(4)
y_pred_itr2
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label="b=0")
plt.plot(x,y_pred_updated,color='blue',label="b=itr1")
plt.plot(x,y_pred_itr2,color='yellow',label="b=itr2")
plt.legend()
plt.show()
```

```python
step_size=le * loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label="b=0")
plt.plot(x,y_pred_updated,color='blue',label="b=itr1")
plt.plot(x,y_pred_itr2,color='blue',label="b=itr2")
plt.plot(x,y_pred_itr3,color='yellow',label="b=itr3")
plt.legend()
plt.show()
```

### Make it as a loop

```python
b=100
m=93.15
le=0.1
epoch=10

for i in range(epoch):
    loss_slope=-2* np.sum(y - m * x.ravel() - b)
    step_size=le * loss_slope
    b=b-step_size
    y_pred=(m*x + b).reshape(4)
    plt.plot(x,y_pred)
    
plt.scatter(x,y)    
print(b)
```
