### 10. converters

```python
def rename(name):
    if name=='Kolkata Knight Riders':
        return 'KKR'
    else:
        return name
```

```python
df=pd.read_csv(r"../Datasets/matches.csv",parse_dates=['date'],converters={'team1':rename,'team2':rename})
df.head()
```
