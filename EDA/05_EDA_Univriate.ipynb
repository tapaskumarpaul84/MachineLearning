```python
import pandas as pd

df=pd.read_csv(r"../Datasets/titanic_train.csv")
```

```python
df.head()
```

### Remove some columns

```python
df=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
df.head()
```

### Identify categorical cols

```python
df.columns
```

```python
object_categorical=[col for col in df.columns if df[col].dtype=='object']
```

```python
object_categorical
```

```python
numeric_categorical=[col for col in df.columns if df[col].dtype!='object' and len(df[col].unique())<10]
```

```python
numeric_categorical
```

```python
categorical_cols=object_categorical+numeric_categorical
```

```python
categorical_cols
```

```python
df['Survived'].value_counts()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Survived',data=df)
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
for col in categorical_cols:
    plt.title(f"Pie plot for column {col}")
    df[col].value_counts().plot(kind='pie',autopct='%.2f')
    plt.figure(figsize=(6,4))
    plt.tight_layout()
    plt.show()
```

### Numerical columns

```python
numercal_columns=[col for col in df.columns if df[col].dtype!='object' and len(df[col].unique())>10]
```

```python
numercal_columns
```

### Histogram

```python
for col in numercal_columns:
    plt.hist(df[col])
    plt.title(f"Histogram plot for column {col}")
    plt.xlabel({col})
    plt.ylabel('Count')
    plt.show()
```

### Distribution plot

```python
for col in numercal_columns:
    sns.displot(df[col],kind='kde')
    plt.title(f"Histogram plot for column {col}")
    plt.xlabel({col})
    plt.ylabel('Count')
    plt.show()
```

### Boxplot

```python
for col in numercal_columns:
    sns.boxplot(df[col])
    plt.title(f"Histogram plot for column {col}")
    plt.xlabel({col})
    plt.ylabel('Count')
    plt.show()
```
